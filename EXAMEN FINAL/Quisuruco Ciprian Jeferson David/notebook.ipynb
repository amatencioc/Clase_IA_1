```python
import pandas as pd
```

```python
file_path = "Absenteeism-data.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
#Eliminamos la columna ID
df = df.drop(columns=["ID"])
df.head()
```

```python
# Variables dummy
reason_dummies = pd.get_dummies(df['Reason for Absence'], prefix='Grupo', drop_first=False)
# Grupo 1:  1-14
df['Grupo_1'] = reason_dummies.loc[:, 'Grupo_1':'Grupo_14'].max(axis=1).astype(int)
# Grupo 2:  15-17
df['Grupo_2'] = reason_dummies.loc[:, 'Grupo_15':'Grupo_17'].max(axis=1).astype(int)
# Grupo 3:  18-21
df['Grupo_3'] = reason_dummies.loc[:, 'Grupo_18':'Grupo_21'].max(axis=1).astype(int)
# Grupo 4:  22-28
df['Grupo_4'] = reason_dummies.loc[:, 'Grupo_22':'Grupo_28'].max(axis=1).astype(int)

# Eliminamos la columna Reason for Absence
df = df.drop(columns=['Reason for Absence'])

df.head()
```

```python
#Extraemos  el mes y dia de la semana de la columan date
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df['Month'] = df['Date'].dt.month
df['Day of the Week'] = df['Date'].dt.weekday  # Lunes=0, Domingo=6
#Eliminamos la columna Date
df = df.drop(columns=['Date'])
df.head()
```

```python
#Convertimos en binario  los datos de la columna education
df['Education'] = df['Education'].apply(lambda x: 1 if x == 0 else 0)
df.head()
```

```python
#Guardamos en un nuevo csv la data modificada
output_path = "df-cleaned.csv"
df.to_csv(output_path, index=False)
print("Archivo guardado en:", output_path)
```
