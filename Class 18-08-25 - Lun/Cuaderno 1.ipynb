```python
#Create funcion drop data:
def FilDF(df,atributo,i,j):
    f1 = df[atributo].between(i,j) 
    df_f = df[f1]
    return df_f
```

```python
#Create funcion retorna dummies:
def GroupDF(df,atributo):
    # Crear variables dummy SOLO para la columna 'Reason for Absence'
    reason_dummies = pd.get_dummies(df[atributo], prefix='Reason') #Dummy value as Reason (motivo)

    # Ordenar columnas para que vayan de 0 a 14
    df_with_dummies=pd.concat([df[['ID']], reason_dummies], axis=1)
    # Agrupar por ID y sumar (cu치ntas veces tuvo cada raz칩n)
    #df_grouped = df_with_dummies.groupby('ID').sum()
    return df_with_dummies.groupby('ID').sum()
```

```python
# Filtrar razones entre 0 y 14 g1 i+14/+3/+4/+7
#Filtrar para 15:17 g2/ 18 :21 g3/ 22:28 g4
df_1=FilDF(data_1,'Reason for Absence',0,14)
print(f"=== Group {1} ===")
GroupDF(df_1,'Reason for Absence')
```
